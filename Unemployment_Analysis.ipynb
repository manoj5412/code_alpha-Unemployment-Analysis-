```python
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
```

```python
df1["Date"] = pd.to_datetime(df1["Date"], dayfirst=True, errors="coerce")
df2["Date"] = pd.to_datetime(df2["Date"], dayfirst=True, errors="coerce")
```

```python
data = pd.concat([df1, df2], ignore_index=True)
data = data.sort_values(by="Date")
```

```python
print(data[["Estimated Unemployment Rate (%)",
            "Estimated Employed",
            "Estimated Labour Participation Rate (%)"]].describe())
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x="Date", y="Estimated Unemployment Rate (%)", data=data, ci=None)
plt.axvspan(pd.to_datetime("2020-03-01"), pd.to_datetime("2020-09-01"),
            color="red", alpha=0.2, label="Covid-19 Lockdowns")
plt.title("Unemployment Rate Trends in India (2019-2020)")
plt.ylabel("Unemployment Rate (%)")
plt.xlabel("Date")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
region_avg = data.groupby("Region")["Estimated Unemployment Rate (%)"].mean().sort_values()
sns.barplot(x=region_avg.values, y=region_avg.index)
plt.title("Average Unemployment Rate by Region (2019-2020)")
plt.xlabel("Unemployment Rate (%)")
plt.show()
```
